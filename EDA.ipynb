### Moving Average

```python
def moving_average(df, compare='Close_x', window=10, min_periods=0):
    df['Moving_avg']= df[compare].rolling(window=window,min_periods=min_periods).mean()
    
    # plot
    rcParams['figure.figsize'] = 12, 4  # width 12, height 4
    ax = df['Moving_avg'].plot(style='b-', grid=True)
    ax = df[compare].plot(style='r-', grid=True)
```

```python
moving_average(df, window=5, min_periods=0)
```

```python
moving_average(df, window=10, min_periods=0)
```

```python
moving_average(df, window=20, min_periods=0)
```
